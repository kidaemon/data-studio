```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## Male Labour Force

```python
df_male = pd.read_excel('lt03-03.xls',sheet_name=1,header=None,usecols='B:P',skiprows=9,skipfooter=39)
df_male.head()
```

```python
df_columns = pd.read_excel('lt03-03.xls',sheet_name=1,header=None,usecols='B:P',skiprows=5)
columns_list = df_columns.iloc[0,:15].tolist()
columns_list[0] = 'Year'
columns_list[13] = '65～69'
columns_list[14] = '70歳以上'
columns_list
```

```python
df_male.columns = columns_list
df_male.head()
```

```python
df_male = df_male.drop(columns=['65歳以上'])
df_male.columns = df_male.columns.str.replace('総数','Total')
df_male.columns = df_male.columns.str.replace('歳','')
df_male.columns = df_male.columns.str.replace('以上','～')
df_male.columns = df_male.columns.str.replace('～','~')
df_male.head()
```

## Female Labour Force

```python
df_female = pd.read_excel('lt03-03.xls',sheet_name=2,header=None,usecols='B:P',skiprows=9,skipfooter=39)
df_female.head()
```

```python
df_columns = pd.read_excel('lt03-03.xls',sheet_name=2,header=None,usecols='B:P',skiprows=5)
columns_list = df_columns.iloc[0,:15].tolist()
columns_list[0] = 'Year'
columns_list[13] = '65～69'
columns_list[14] = '70歳以上'
columns_list
```

```python
df_female.columns = columns_list
df_female.head()
```

```python
df_female = df_female.drop(columns=['65歳以上'])
df_female.columns = df_female.columns.str.replace('総数','Total')
df_female.columns = df_female.columns.str.replace('歳','')
df_female.columns = df_female.columns.str.replace('以上','～')
df_female.columns = df_female.columns.str.replace('～','~')
df_female.head()
```

```python
plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'axes.edgecolor': '#ffffff',
    'xtick.major.size': 3,
    'xtick.minor.size': 0,
    'ytick.major.size': 0,
    'ytick.minor.size': 0,
    'legend.edgecolor': '#ffffff',
    'ps.fonttype': 42,
})
```

```python
for i in range(0,51):
    y = df_male.columns[2:]
    x1 = df_male.iloc[i,2:]
    x2 = df_female.iloc[i,2:]

    fig, axes = plt.subplots(ncols=2, sharey=True)
    axes[0].barh(y, x1, align='center', color='#01579B')
    axes[1].barh(y, x2, align='center', color='#B71C1C')
    axes[0].invert_xaxis()
    axes[0].set_xticks([0, 150, 300, 450, 600])
    axes[1].set_xticks([0, 150, 300, 450, 600])
    axes[0].yaxis.tick_right()
    axes[0].set(title=df_male.Year[i])
    fig.tight_layout()
    fig.subplots_adjust(wspace=0.2)
    plt.savefig('images/graph-pyramid-'+str(i)+'.png')
    plt.show()
```

# Total Labour Force and Average Age

```python
plt.bar(df_male.Year,df_male.Total)
plt.bar(df_female.Year,df_female.Total,bottom=df_male.Total)
plt.show()
```

```python
for i in range(0,51):
    for j in range(2,14):
        df_male.iloc[i,j] = df_male.iloc[i,j] / df_male.iloc[i,1]
```

```python
df_male.head()
```

```python
avg_age = []
for i in range(0,12):
    age = 17.5 + 5 * i
    avg_age.append(age)
avg_age
```

```python
avg_list = []
for i in range(0,51):
    avg = sum(df_male.iloc[i,2:14]*avg_age)
    avg_list.append(avg)
```

```python
df_male['avg_age'] = avg_list
df_male.head()
```
