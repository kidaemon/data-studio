## Why are there so few female surgeons in Japan?

Early August, a scandal exposed at one of the largest medical school in Japan. Tokyo Medical University admitted that they discriminated females at the entrance exams for more than decades. One of the reasons is job mismatch in medical sector. Female doctors tend to choose internal medicine for their jobs and the shortage of surgeons is a serious problem. Why is it hard for women to be surgeons?

https://www.mhlw.go.jp/toukei/saikin/hw/ishi/16/index.html

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pd.set_option('max_colwidth',-1)
department = pd.read_excel('toukeihyo_H28.xls', sheet_name='統計表10',skiprows=5,encoding='shift_jis')
department = department.dropna(axis=1, how='all')
department = department.iloc[:-3,:4]
department.columns = ['Department','Total','Male','Female']
department.Department = pd.read_csv('department.csv')
department
```

```python
department['Male%'] = department.Male / department.Total * 100
department['Female%'] = department.Female / department.Total * 100
department.sort_values(by='Male%')
```

```python
plt.rcdefaults()
plt.rcParams.update({
    'pdf.fonttype': 42,
    'axes.grid': True,
    'axes.edgecolor': '#ffffff',
    'xtick.major.size': 0,
    'xtick.minor.size': 0,
    'ytick.major.size': 0,
    'ytick.minor.size': 0,
    'legend.edgecolor': '#ffffff',
    'ps.fonttype': 42,
})
```

```python
ax = department[['Department','Male%','Female%']].sort_values(by='Male%').plot(kind='barh',x='Department',stacked=True,figsize=(8, 10))
ax.set_title('Surgery is dominated by male doctors', pad=40, fontsize=20)
ax.legend(ncol=2,loc='upper center',bbox_to_anchor=(0.5, 0.92),bbox_transform=plt.gcf().transFigure,frameon=False)
ax.grid(False, axis='y')
ax.set_xlim((0, 100))
plt.show()
```
